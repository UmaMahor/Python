# Creating a Function

##### In Python a function is defined using the def keyword:

```python
def email_func(name):
    print(name.replace(" ","")+'123@monstersinc.com')
```

# Calling a Function

##### To call a function, use the function name followed by parenthesis:

##### Company name is monsters.inc. Create corporate email id when user input name

```python
email_func("Mike Wazowski")
email_func("James P Sullivan")
email_func("Boo")
```
