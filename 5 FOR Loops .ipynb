# Nested For Loops


```python
flavors = ['Vanilla', 'Chocolate', 'Cookie Dough']
toppings = ['Hot Fudge', 'Oreos', 'Marshmallows']
```

```python
for v1 in flavors:
    for v2 in toppings:
        print(v1, 'topped with', v2)
```
