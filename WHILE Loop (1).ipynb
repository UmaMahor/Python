```python
c= 0
while c<5:
    print(c,'is less than 5')
    c=c+1
```

##### Break statement can break the loop even if while condition is true

```python
c=0
while c<5:
    print(c,'is less than 5')
    if c==3:
        break
    c=c+1
```

```python
c=0
while c<5:
    print(c,'is less than 5')
    if c==3:
     break
    c=c+1
else:
    print('Number is not less than 5')
    
```

##### Continue

##### when c reaches 3 it will not break the loop.It will continue to while c<5 loop

```python
c=0
while c<5:
    c=c+1
    if c==3:
     continue
    print(c,'is lessthan5')  
else:
    print('No longer less than 5')
    
```
